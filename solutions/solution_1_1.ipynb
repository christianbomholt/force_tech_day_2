  - Filter out all non-French wines
  - group by title 
  - create a column with average points
  - print the top 6 wines
  - Are these the wines you would choose? Try to improve top 6

```python
import pandas as pd

wines = pd.read_csv("../data/winemag-data-130k-v2.csv")
wines.head(2)
```

```python
wines.where(wines.country=="France")\
    .groupby(["title","country"])\
    .mean()\
    .sort_values(['points','price'], ascending=[False, True])\
    .dropna()\
    .head(6)
```

```python
french_wines = wines.where(wines.country=="France")\
    .groupby(["title","country"])\
    .mean()\
    .dropna()

french_wines['value_for_money'] = french_wines.points/french_wines.price 
```

```python
french_wines.where(french_wines.points>90).sort_values("value_for_money",ascending = False).dropna()
```

## Alternative


```python
french_wines = wines.query("country=='France'")

french_wine_mean = french_wines[["title","points","price","province"]].groupby("title").mean().dropna()
french_wine_mean
```

```python
french_wine_mean[["price","points"]].sort_values(["points","price"],ascending=[False,True]).head(6)
```

```python

french_wine_mean.loc[:,"value_for_money"] = french_wine_mean["points"]/french_wine_mean["price"] 

french_wine_mean.sort_values("value_for_money", ascending=False).query("points>90")
```
