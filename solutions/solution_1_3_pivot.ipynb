````python
import pandas as pd 

sp500 = pd.read_csv("../data/sp500.csv", index_col = "Date")
sp500.head()
````

````python
import plotly.express as px
df = sp500.loc[:,["Low","High"]].reset_index().melt(id_vars="Date")

px.line(df,x="Date",y="value",color="variable")
````
