$$
    \text{AveragePrice} = \alpha + \beta \cdot \text{Total Volume} + \varepsilon, \quad \varepsilon \sim N(0,1)
$$

````python
import pandas as pd
from sklearn import linear_model
import plotly.graph_objects as go
import plotly.express as px



avocados = pd.read_csv("../data/avocado.csv", usecols=["Date","AveragePrice","total_volume","region","type"])\
.query('region=="Albany"').query('type=="organic"').drop(["region","type"],axis=1)
avocados.head()
````

````python
reg = linear_model.LinearRegression()

X = avocados.total_volume.values.reshape(-1, 1)
y = avocados.AveragePrice.values
reg.fit(X,y)
````

````python
fig = go.Figure()
x= avocados.total_volume.values
line1 = reg.coef_[0]*x + reg.intercept_
# Add traces

fig.add_trace(go.Scatter(x=x, y=y,mode='markers',name='observations'))
fig.add_trace(go.Scatter(x=x, y=line1,mode='lines',name='OLS formula'))
````

````python
avocados["pred"] = reg.coef_[0]*X + reg.intercept_
# Add traces
px.line(avocados.drop("total_volume",axis=1).melt(id_vars="Date"), x="Date", y="value", color="variable")
````
